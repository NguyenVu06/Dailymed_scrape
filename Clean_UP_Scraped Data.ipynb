```python
df = pd.read_csv('Propofol_21Apr21.csv', header=None)
```

```python

df.columns= ["type", "route", "API", "Concentration", "Inactive_1","i1C",  "Inactive_2", "i2C", "Inactive_3","i3C", "Inactive_4","i4C",  "Inactive_5", "i5C", "Inactive_6","i6C", "Inactive_7", "DESC"]
```

```python
df.head()
```

```python
c_names = ["type", "route", "API", "Concentration", "Inactive_1","i1C",  "Inactive_2", "i2C", "Inactive_3","i3C", "Inactive_4","i4C",  "Inactive_5", "i5C", "Inactive_6","i6C", "Inactive_7", "DESC"]

for c in c_names:
    df[c] = df[c].str.upper()
    df[c]=df[c].str.replace(r"\(([^\)]+)\)", "")
    

df["excipients"]=df[["Inactive_1","Inactive_2","Inactive_3","Inactive_4","Inactive_5","Inactive_6", "Inactive_7"]].values.tolist()
df["excipients"] = df["excipients"].apply(lambda x: [v for v in x if v == v])
df["excipients"] = df["excipients"].apply(lambda x: sorted(x))
df["excipients"] = df["excipients"].apply(lambda x: ",".join(x))
```
