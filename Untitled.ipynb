```python
import os
import pandas as pd
import numpy as np
import regex as re
```

```python
rootpath = r"C:\Users\nvu\Downloads"
folder_name = input("Enter file name")
path = os.path.join(rootpath,folder_name)
```

```python
data = pd.read_csv(path)
data.head(3)
data.replace(np.nan," ")
data_excipients = data.iloc[:,[i*2 for i in range(9)]]
```

```python
cols = data_excipients.columns
for c in cols:
    data_excipients[c].fillna("NA")
    data_excipients[c].astype(str)
    
```

```python
df_excipients_np = data_excipients.iloc[:, 2:].values
```

```python
print(type(np.nan))
```

```python
stored = []
for row in df_excipients_np:
    r = row[~pd.isnull(row)]
    r = (",".join(sorted((list(r))))).lower()
    r = re.sub(r'\([^)]*\)', '', r)
    stored.append(r)
```

```python
data["Excipients"] = stored 
```

```python
data.to_csv(os.path.join(os.getcwd(),"Cipro.csv"))
```
